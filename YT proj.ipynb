# Generating YouTube DATA using Google API

```python
from googleapiclient.discovery import build
import numpy as np
import pandas as pd
```

Creating Function to generate the Data

```python
def YT(cat):
    req = service.videos().list(part='statistics,snippet', chart='mostPopular',videoCategoryId=cat)
    res = req.execute()
    title = []
    Category = []
    channel = []
    Views = []
    Likes = []
    dislikes = []
    comments = []
    while req:
        res = req.execute()
        for k in res["items"]: 
            title.append(k['snippet']['title'])
            Category.append(int(k['snippet']['categoryId']))
            channel.append(k['snippet']['channelTitle'])
            Views.append(int(k["statistics"]['viewCount']))
            try:
                Likes.append(int(k["statistics"]['likeCount']))
            except: 
                Likes.append(np.nan)
            try:
                dislikes.append(int(k["statistics"]['dislikeCount']))
            except:
                dislikes.append(np.nan)
            try:
                comments.append(int(k["statistics"]['commentCount']))
            except:
                comments.append(np.nan)
        req = service.search().list_next(req, res)
    da = pd.DataFrame({ 'Title': title,
                    'Channel' : channel,
                    'Views': Views,
                    'Likes': Likes,
                    'Dislikes' : dislikes,
                    'Comments' : comments})  
    return da
```

##    Requesting Data for the following Catgories

    20 -- Gaming

```python
da_music = YT(10)
da_music["Category"] = 'Music'
da_gaming = YT(20)
da_gaming["Category"] = 'Gaming'
```
